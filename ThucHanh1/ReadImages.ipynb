```python
#include libraries
import cv2 as cv
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import glob

def load_image(image_path):
    coloured_image = cv2.imread(image_path)
    grey_image = cv2.cvtColor(coloured_image, cv2.COLOR_BGR2GRAY)
    print('image matrix size: ', grey_image.shape)    
    return grey_image
#Load an image in the directory
img=cv.imread('images/Bk2017.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize = (10,10))
plt.imshow(img)
plt.show()
#Convert color to gray image
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure(figsize = (10,10))
plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
plt.show()
cv.imwrite('Bk2017_gray.jpg',gray)
#Get a region in the image
ROI = gray[:5, :5]
print(ROI)
plt.figure(figsize = (1,1))
plt.imshow(ROI)
plt.show()
#Get image information
print img.shape
print img.size
print gray.shape
print gray.size
print ROI.shape
print ROI.size      
```
